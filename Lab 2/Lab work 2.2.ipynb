### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел генеруємо рандомно  
генеруємо тисячу чисел в діапазоні від 0 до 10

```python
import random
lwrn = [random.randint(0,10) for i in range(1000)]

# Максимум
def exmax(l):
    n = l[0]
    for number in l:
        if number > n:
            n = number 
    return n 

# Мінімум
def exmin(l):
    n = l[0]
    for number in l:
        if number < n:
            n = number 
    return n 

# Середнє значення
def exava(l):
    sumpr = 0
        
    for number in l:
        sumpr += number
        av = sumpr / len(l) 
    return av 

# Медіана
def exmed(l):
    l1 = sorted(l)
    
    if len(l1) % 2:
        p = len(l1)//2
        return l1[p]

    else:
        pl  = (len(l1)-1) // 2
        pr = (len(l1)+1) // 2
        return (l1[pl] + l1[pr]) / 2

# Частота
def exf(l):
    f = {}
    
    for number in l:
        c = f.get(number,0)
        f[number] = c + 1
        
    return f


# Середнє квадратичне відхилення
def exad(l):
    sumd1 = 0
    sumd2 = 0
        
    for number in l:
        sumd1 += number
        avd1 = sumd1 / len(l)
        sumd2 += number**2
        avd2 = sumd2 / len(l)
        ad = (avd2 - avd1**2)**(1/2)
        
    return ad


print('Максимальне значення: ' + str(exmax(lwrn)))
print('Мінімальне значення: ' + str(exmin(lwrn)))
print('Середнє значення: ' + str(round(exava(lwrn), 1)))
print('Медіана: ' + str(exmed(lwrn)))
print('Частота: ' + str(exf(lwrn)))
print('Середнє квадратичне відхилення: ' + str(round(exad(lwrn), 3)))
```
